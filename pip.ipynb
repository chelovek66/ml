```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.metrics import log_loss, matthews_corrcoef, cohen_kappa_score
from sklearn.metrics import balanced_accuracy_score, zero_one_loss, hinge_loss
from sklearn.metrics import jaccard_score, hamming_loss, precision_recall_curve

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

# Classification Report (все метрики сразу)
report = classification_report(y_true, y_pred)

# ROC-AUC (для бинарной классификации)
auc = roc_auc_score(y_true, y_pred_proba)

# Log Loss (логарифмическая потеря)
log_loss_value = log_loss(y_true, y_pred_proba)

# Matthews Correlation Coefficient (-1 до +1)
mcc = matthews_corrcoef(y_true, y_pred)

# Cohen's Kappa (согласие между предсказаниями)
kappa = cohen_kappa_score(y_true, y_pred)

# Balanced Accuracy (для несбалансированных классов)
balanced_acc = balanced_accuracy_score(y_true, y_pred)

# Zero-One Loss (0-1 потеря)
zero_one = zero_one_loss(y_true, y_pred)

# Hinge Loss (для SVM)
hinge = hinge_loss(y_true, y_pred)

# Jaccard Score (похожесть множеств)
jaccard = jaccard_score(y_true, y_pred, average='binary')

# Hamming Loss (доля неправильных предсказаний)
hamming = hamming_loss(y_true, y_pred)

# Precision-Recall Curve
precision_curve, recall_curve, _ = precision_recall_curve(y_true, y_pred_proba)
```

```python
from sklearn.metrics import mean_squared_log_error, mean_absolute_percentage_error
from sklearn.metrics import median_absolute_error, explained_variance_score
from sklearn.metrics import max_error, mean_gamma_deviance, mean_poisson_deviance

# MSLE (Mean Squared Logarithmic Error)
msle = mean_squared_log_error(y_true, y_pred)

# MAPE (Mean Absolute Percentage Error)
mape = mean_absolute_percentage_error(y_true, y_pred)

# MedAE (Median Absolute Error)
medae = median_absolute_error(y_true, y_pred)

# Explained Variance Score
explained_var = explained_variance_score(y_true, y_pred)

# Max Error (максимальная ошибка)
max_err = max_error(y_true, y_pred)

# Mean Gamma Deviance (для гамма-регрессии)
gamma_dev = mean_gamma_deviance(y_true, y_pred)

# Mean Poisson Deviance (для пуассоновской регрессии)
poisson_dev = mean_poisson_deviance(y_true, y_pred)
```
